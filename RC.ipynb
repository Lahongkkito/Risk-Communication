```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

# dataframe의 행이 잘리는 것을 방지
pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',200)
```

```python
rc = pd.read_csv("MYRC.csv")
rc.head()
```

```python
rc.info()
```

```python
rc.describe()
```

```python
# 결측값 처리

## time_2의 null값은 "오후"라고만 기록되어 있어 오후에 해당하는 time_2 >= 0.5의 평균값 사용
afternoon = rc[rc["time_2"] >= 0.5]
mean_time_2 = afternoon["time_2"].mean()
rc["time_2"].fillna(mean_time_2, inplace=True)

## BPA_2는 0값이 존재, 검출한계/2에 해당하는 값(0.0075)으로 대체
rc["BPA_2"].replace(0, 0.0075, inplace=True)

## MEOHP_1 missing은 하나의 데이터이므로 제거
rc = rc.dropna()
```

```python
# 나이 column 생성

## 이상치 확인 - 입력 오류 확인 후 실제 값 대체
rc.loc[19,'birth']=19590523

def trans_age(birth):
    age = 2021 - birth//10000
    return age
    
rc["age"] = rc["birth"].map(lambda x: trans_age(x))
```

```python
## 검출값들의 이상치 확인 - 결과 이상자이므로 제거 

rc.sort_values(by=["BPA_1"], ascending=False).head(10)
# rc.sort_values(by=["MnBP_1"], ascending=False).head()
# rc.sort_values(by=["MEOHP_1"], ascending=False).head()
# rc.sort_values(by=["MEHHP_1"], ascending=False).head()
# rc.sort_values(by=["BPA_2"], ascending=False).head()
# rc.sort_values(by=["MnBP_2"], ascending=False).head()
# rc.sort_values(by=["MEOHP_2"], ascending=False).head()
# rc.sort_values(by=["MEHHP_2"], ascending=False).head()

rc.drop([23, 84, 174], inplace=True)    # 2017 국민환경보건기초조사 기준 참고
rc.reset_index(drop=True)
```

```python
# sns.heatmap(rc.corr())
```

```python
# 결과값의 상관관계와 분포 확인

horm_1 = ["BPA_1", "MnBP_1", "MEOHP_1", "MEHHP_1"]
horm_2 = ["BPA_2", "MnBP_2", "MEOHP_2", "MEHHP_2"]
horm_pair = zip(horm_1, horm_2)
horms = horm_1 + horm_2

for i, j in horm_pair:
    print(np.corrcoef(rc[i], rc[j])[1,0])
```

```python
fig, axes = plt.subplots(2,4, figsize=(15,8))

for i, (ax, idx) in enumerate(zip(axes.flatten(), horms)):
    sns.distplot(rc[idx], ax=ax)
```

```python
## log변환 후 분포 확인

log_horms = []
for i in horms:
    name = "log_{}".format(i)
    log_horms.append(name)
    rc[name] = rc[i].map(lambda x : np.log(x))

    
fig, axes = plt.subplots(2,4, figsize=(15,8))
colors = ['tab:blue', 'tab:red', 'tab:olive','tab:purple']

for i, (ax, idx) in enumerate(zip(axes.flatten(), log_horms)):
    sns.distplot(rc[idx], ax=ax, color=colors[i%4])
```
